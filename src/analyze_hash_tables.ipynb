```kotlin
%use dataframe, lets-plot
```

```kotlin
val collisionsDataFrame = DataFrame.readCsv("../collisions.csv")
collisionsDataFrame
```

```kotlin
val cols = collisionsDataFrame.toMap()
val plot = letsPlot(cols) {
    x = "index"
    weight = "collisions"
} + geomBar(color = "blue")
plot
```

```kotlin
val sizeDf = DataFrame.readCsv("../sizes.csv")
sizeDf
```

```kotlin
val sizes = sizeDf.toMap()
val plot = letsPlot(sizes) {
    x = "index"
    weight = "size"
} + geomBar(color = "#FA8072")
plot
```

```kotlin
val sizeCollsDf = collisionsDataFrame.join(sizeDf){ index match right.index  }
sizeCollsDf
```

```kotlin
val plot = letsPlot(sizeCollsDf.toMap()) {
    x = "index"
    y = "collisions"
} + geomBar(color = "blue", fill="blue", manualKey = "Моё хеширование") { weight = "collisions" } + geomBar(
    color = "#FA8072", fill = "#FA8072",
    manualKey = "Бин-ное хеширование",
) {
    weight = "size"

} + guides()
plot
```
